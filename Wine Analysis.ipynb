## Checking the Correlation between the different columns

## Alcohol vs Wine Quality

## Volatile Acidity vs Alcohol

#### Volatile acidity (acetic acid) contributes to the overall flavour and smell of the wine. 
#### Below is a plot to illustrate the relationship between volatile acidity and Wine Quality

```python
sns.lineplot(y='volatile acidity', x='quality', data=data)
plt.show()
sns.barplot(y='volatile acidity', x='quality', data=data)
```

#### Suprisingly, the above graph proves that the higher the volatile acidity, the less the quality of the wine

```python
data.sort_values(['alcohol','quality'])
```

#

```python
sorted_volatile_acidity = data[['volatile acidity','quality']].sort_values(['volatile acidity','quality'], ascending = [0,1])
sorted_volatile_acidity = sorted_volatile_acidity.drop_duplicates()
sorted_volatile_acidity
```

#### According to the sorted dataframe above it can be observed that the higher the volatile acidity, the less the quality of the wine. It can be visualized using the graphs plotted below

```python
sns.lineplot(y='volatile acidity', x='quality', data=sorted_volatile_acidity)
plt.xlabel('Quality')
plt.ylabel('Volatile Acidity')
```
